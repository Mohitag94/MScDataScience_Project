```python
path = r"D:\MScDataScience\7.Data_Science_Project\SourceCode\Plots"
title="single"
model=keras.saving.load_model(r"D:\MScDataScience\7.Data_Science_Project\Model_Callbacks\Single_lstm\hp_96Batch_50epochs.keras")
# saving overall view of model
plot_model(model,
            to_file=os.path.join(path,
                                f"{title}_seq.png"),
            show_shapes=True,
            show_layer_names=True)
# saving layered view
visualkeras.layered_view(model,
                            legend=True,
                            to_file=os.path.\
                            join(path,
                                f"{title}_layered_view.png"), scale_z=0.01, scale_xy=4)
```
